**A diferencia de las otras columnas con valores nulos, la mayoria de los valores nulos de la columna job_industry_category (categoria de trabajo en la industria) se pueden imputar por una de las categorias que se han registrado en otros clientes. Esto gracias a la columna job_title (cargo laboral), ya que existen trabajos en los clientes cuya categoria es nula, que se repiten en otros clientes y que si poseen una categoria de trabajo en la industria.**

Por ejemplo:

Existen clientes que tienen un cargo laboral de "Account Executive" pero no poseen un valor en la columna job_industry_category, por otro lado existen clientes que desempeñan el mismo cargo y que si poseen job_industry_category:

    (job_title)         (job_industry_category)   (cantidad de registros)

    Account Executive	Entertainment	                    1

    Account Executive	Financial Services	                1

    Account Executive	Health	                            2

    Account Executive	IT	                                1 

    Account Executive	Manufacturing                       3	
    
Pero que categoria elegir para la imputación, es aqui donde interviene la moda, para el ejemplo propuesto la mejor opcion seria imputar el valor nulo por la categoria "Manufacturing".


```python
df_nuevos_clientes.job_industry_category.value_counts(dropna=False, normalize=True)*100
```

```python
# Obtenemos una lista de los cargos laborales (job_title) de los clientes que no poseen una categoria de trabajo en la industria (job_industry_category)
# Cabe resaltar que no se toman en cuenta aquellos clientes que a parte de no tener una categoria de trabajo, no poseen un cargo laboral.
lista_trabajos_sin_categoria = list(df_nuevos_clientes[(df_nuevos_clientes.job_industry_category.isnull()) & (df_nuevos_clientes.job_title !='Desconocido')].job_title)
lista_trabajos_sin_categoria
```

```python
# Creamos un dataframe con los clientes que desempeñan un cargo laboral igual a los que se encuentran en la lista anteriormente
# creada y que si poseen una catgoria de trabajo en la industria
df_clientes_trabajo_con_categoria = df_nuevos_clientes[(df_nuevos_clientes.job_title.str.contains('|'.join(lista_trabajos_sin_categoria))) & (df_nuevos_clientes.job_industry_category.isnull() == False)]
df_clientes_trabajo_con_categoria
```

```python
#Creamos un dataframe con el numero de ocurrencias por cada grupo de cargo laboral (job_title) y categoria de trabajo en la industria (job_industry_category)
df_ocurrencias_trabajo_categoria = df_clientes_trabajo_con_categoria.groupby(['job_title','job_industry_category']).size().reset_index(name='cantidad')
df_ocurrencias_trabajo_categoria
```

```python
# Creamos un dataframe de los grupos cargo laboral (job_title) y categoria de trabajo en la industria (job_industry_category) con mayor ocurrencia
# y lo convertimos en un diccionario
df_moda_trabajo = df_ocurrencias_trabajo_categoria.loc[df_ocurrencias_trabajo_categoria.groupby(['job_title']).cantidad.idxmax()].reset_index(drop=True)
df_moda_trabajo
```

```python
diccionario_trabajos_categoria = df_moda_trabajo.set_index('job_title')['job_industry_category'].to_dict()
diccionario_trabajos_categoria
```

```python
# Creamos un dataframe con los clientes que no poseen una categoria de trabajo en la industria e imputamos los valores nulos
# mapeando los cargos laborales en el diccionario anteriormente creado, es de suma importancia que los registros conserven
# el indice de origen para su posterior reincorporación al dataframe original.
df_categoria_nulos = df_nuevos_clientes[(df_nuevos_clientes.job_industry_category.isnull()) & (df_nuevos_clientes.job_title != 'Desconocido')]
df_categoria_nulos.job_industry_category = df_categoria_nulos.job_title.map(diccionario_trabajos_categoria)

df_categoria_nulos
```

```python
# Podemos observar que despues de la imputación solo nos quedan 6 clientes que no cuentan con una categoria de trabajo pero que
# si tienen un cargo laboral conocido
df_categoria_nulos.job_industry_category.value_counts(dropna=False)
```

```python
# Reincorporamos los registros imputados al dataframe original.
df_nuevos_clientes.update(df_categoria_nulos)
```

```python
# Al final son 28 clientes que no poseen un valor en la columna job_industry_category, entre los que si tienen un cargo laboral
# conocido y los que no se conoce su ocupación. Estos 28 valores nulos los imputaremos con el valor "Desconocido"
df_nuevos_clientes.job_industry_category.value_counts(dropna=False)
```

**Es en este punto donde haremos uso de una API que nos ayudara a determinar los generos de los clientes por su nombre. Esta API es proporcionada por genderize.io**

```python
# Primero obtendremos una lista con los nombres de aquellos clientes que no poseen un genero
lista_nombres_sin_genero = df_nombres_sin_genero[df_nombres_sin_genero.gender.isnull()].first_name.to_list()
lista_nombres_sin_genero
```

```python
# Creamos una función que consistira principalmente en consultar cada nombre de una lista ingresada a la api y a la vez agregara
# la respuesta proporcionada a un diccionario, cuyas claves seran los nombres y los valores, los generos. La función retornara
# dicho diccionario.

def obtenerGenero(lista_nombres):
    nombres_con_genero = {}
    consulta = "https://api.genderize.io?name="
    
    for nombre in lista_nombres:
        resultado = requests.get(consulta + nombre)
        respuesta = resultado.json()
        if respuesta["gender"] != None:
            if respuesta["gender"] == "male":
                nombres_con_genero[respuesta['name']] = "Male"
            else:
                nombres_con_genero[respuesta['name']] = "Female"
    
    return nombres_con_genero
```

```python
# El diccionario retornado por la función se agregara al diccionario creado inicialmente para mapear los nombres y obtener los
# generos.
# Remplazamos los valores nulos de la columna gender con el resultado del mapeo entre los nombres del dataframe y el diccionario
resultado = obtenerGenero(lista_nombres_sin_genero)
dict_nombres_con_genero.update(resultado)
df_nombres_sin_genero.gender = df_nombres_sin_genero.first_name.map(dict_nombres_con_genero)
```

```python
# Al final, de los 88 clientes que no poseian genero, solo han quedado 3. A los cuales se les proporcionará el genero segun el
# criterio del analista.
df_nombres_sin_genero[df_nombres_sin_genero.gender.isnull()]
```
